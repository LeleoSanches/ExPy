```python
x = [1, 2, 3]
y = [1, 4, 9]
z = [10, 5, 0]
plt.plot(x, y) # Plota os valores
plt.plot(x, z) # Plota os valores
plt.title("Test Plot") # Nome do gráfico
plt.xlabel("x") # Nome do eixo x
plt.ylabel("y") # Nome do eixo y
plt.legend(["this is y", "This is z"]) #Legenda dos plots
plt.show()
```
